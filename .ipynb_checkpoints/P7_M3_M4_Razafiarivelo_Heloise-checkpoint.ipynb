###  Importez les données

```python
df = pd.read_csv('DATA/DATAP7_db.csv')
df
```

```python
nb_quantiles = 10       # nombre de quantiles (nombre de classes de revenu) 
n  = 1000*nb_quantiles   # taille de l'échantillon, supérieure à 1000 fois le nb de quantile
pj = 0.9 # coefficient d'élasticité

def generate_incomes(n, pj):
    # On génère les revenus des parents (exprimés en logs) selon une loi normale.
    # La moyenne et variance n'ont aucune incidence sur le résultat final (ie. sur le caclul de la classe de revenu)
    ln_y_parent = st.norm(0,1).rvs(size=n)
    # Génération d'une réalisation du terme d'erreur epsilon
    residues = st.norm(0,1).rvs(size=n)
    
    return pd.DataFrame(list(zip(np.exp(pj*ln_y_parent + residues), np.exp(ln_y_parent))))  # revenus des enfants puis revenu des parents

# première colonne : revenu des enfants et deuxième colonne : revenu des parents

incomes = pd.DataFrame(generate_incomes(n, pj))
incomes.columns = ['y_child', 'y_parents']
size = len(incomes)
```

```python
y_child = incomes.y_child
incomes_sorted_child = incomes.sort_values(by='y_child')
sorted_y_child = incomes_sorted_child.y_child

def quantiles(y_child, nb_quantiles):  
    quantiles = np.round(np.arange(1, nb_quantiles+1, nb_quantiles/size) -0.5 +1./size) # on définit des quantiles de 1 à 100
    q_dict = {a:int(b) for a,b in list(zip(sorted_y_child,quantiles))} # dictionnaire, le zip permet de créer un df avec les revenus + quantiles
    return pd.Series([q_dict[e] for e in y_child])

c_i_child = quantiles(y_child, nb_quantiles) # on stocke le dataframe revenu et classe de revenu des enfants

y_parents = incomes.y_parents
inc_copy = incomes.copy()
incomes_sorted_parents = inc_copy.sort_values(by='y_parents')
sorted_y_parents = incomes_sorted_parents.y_parents
def quantiles(y_parents, nb_quantiles): 
    quantiles = np.round(np.arange(1, nb_quantiles+1, nb_quantiles/size) -0.5 +1./size) # on définit des quantiles de 1 à 100
    q_dict = {a:int(b) for a,b in list(zip(sorted_y_parents,quantiles))} # dictionnaire, le zip permet de créer un df avec les revenus + quantiles
    return pd.Series([q_dict[e] for e in incomes['y_parents']])

c_i_parent = quantiles(y_child, nb_quantiles)

def compute_quantiles(y_child, y_parents, nb_quantiles):
        sample = pd.concat([y_child, y_parents, c_i_child, c_i_parent], axis=1) # on concatène l'ensemble des informations
        sample.columns = ["y_child", "y_parents", "c_i_child","c_i_parent"]
        return sample
    
sample = compute_quantiles(y_child, y_parents, nb_quantiles)

def conditional_distributions(sample, nb_quantiles):
    counts = sample.groupby(["c_i_child","c_i_parent"], dropna=False).apply(len).unstack(fill_value=0).stack() # on dénombre le nbe de parents de chaque
    # classe de revenu
    counts = counts.reset_index()
    counts.columns = ["c_i_child","c_i_parent","counts"]
    mat = counts # nombre de parents de chaque classe
       
    return np.array(mat) # on extrait ces informations

cd = conditional_distributions(sample, nb_quantiles)
cd_df = pd.DataFrame(conditional_distributions(sample, nb_quantiles))
cd_df.columns = ["c_i_child","c_i_parent","counts"]
cd_df

def proba_cond(c_i_parent, c_i_child, cd):
    cd_df['nombre'] = 1000
    cd_df['p'] = cd_df['counts']/cd_df['nombre']
    # print(cd_df)
    return cd_df[['p']]

p = proba_cond(c_i_parent, c_i_child, cd)
p

proba_concat = pd.DataFrame(conditional_distributions(sample, nb_quantiles))
proba_concat['p'] = p
proba_concat.columns = ["c_i_child", "c_i_parent", "counts", "p"]
proba_concat
```

__Représentation graphique des distributions conditionnelles :__

Pour obtenir un graphique clair, nous limitons le nombre de quantiles à 10.

```python
def to_plot_conditional_distributions(sample):
    counts = pd.DataFrame(sample)
    crosstab = pd.crosstab(index=counts['c_i_child'], columns=counts['c_i_parent'], normalize='index',margins=False)
    return crosstab
```

```python
tplcd = to_plot_conditional_distributions(sample)
tplcd
```

```python
def plot_conditional_distributions(tplcd):
    tplcd.plot(kind="bar",stacked=True,figsize=(10,8))
    plt.xlabel("Quantile enfant")
    plt.ylabel("Probabilité quantile parent")
    plt.title("Distributions conditionnelles")
    plt.savefig("Distributions_conditionelles.jpg")
```

## Réalisez un test de Fisher

On procède à un **test de Fisher**.µ

Hypothèse H0 : pays sans impact sur le revenu de l'individu.
Hypothèse H1 : le pays a un impact sur le revenu de l'individu.

La p-valeur de ce test (0) est inférieure à 5 %. On rejette donc l'hypothèse H0 selon laquelle $α_{1}=α_{2}=α_{3}=...=α_{i}=0$. 

Le pays a un impact sur le revenu de l'individu.

Pour obtenir le tableau de l'analyse de la variance, on utilise la commande anova_lm :

```python
sm.stats.anova_lm(anova_pays, typ=2)
```

_Quel est le pourcentage de variance expliquée par votre modèle ?_

On utilise la commande  lm  pour régresser le revenu de l'individu, en fonction des variables revenu moyen et indice de Gini.

```python
# On ne garde que les vaariables explicatives et la variable expliquée :
df_reg = df_final[['income', 'mj', 'Gj']]
df_reg
```

```python
reg_multi = smf.ols('income~mj+Gj', data=df_reg).fit()
print(reg_multi.summary())
```

```python
# test au niveau alpha = 5%
alpha = 0.05
```

Le pourcentage de variance expliquée par le modèle est de __49,7%.__
C'est intéressant mais pas suffisant pour expliquer le revenu de l'individu, uniquement sur la base du revenu moyen et du coefficient de Gini.

On trouve pour le coefficient de Gini une p-value de 1, ce qui est très au-dessus du seuil de 5%.
Pour la variable moyenne des revenu, la p-value est bien inférieure au seuil de 5%.
Il n'y a pas d'effet significatif du coefficient de Gini sur le revenu de l'individu. Mais un effet du revenu moyen.

__Selon ce modèle, donnez la décomposition de variance totale expliquée par :__
- Le pays de naissance (ie. le revenu moyen et l’indice de Gini);
- Les autres facteurs non considérés dans le modèle (efforts, chance, etc.)

```python
# fitted model : reg multi
anova_table = sm.stats.anova_lm(reg_multi, typ=2) # incorporate fitted model
anova_table
```

```python
analyse_var_exp=sm.stats.anova_lm(reg_multi, typ=2)
SCE=analyse_var_exp.iloc[0,0]
SCR=analyse_var_exp.iloc[2,0]
mj_var=SCE/(SCE+SCR)
mj_var

print('La variance expliquée par le revenu du moyen est de 47%.')
```

### Calcul des leviers

On peut calculer les leviers comme ceci, en sachant que le seuil des leviers est de $2∗\frac{p}{n}$.

```python
analyse_resultat = log_df2.copy()
analyse_resultat['levier'] = log_reg_multi2.get_influence().hat_matrix_diag

p = 3 # nombre de variables
n = len(log_df2)

seuil_levier = 2*p/n

analyse_resultat
```

```python
seuil_levier
```

```python
plt.figure(figsize=(10,6), dpi=300)
plt.bar(analyse_resultat['country'], analyse_resultat['levier'])
plt.xticks(np.arange(0, 12000, step=3), fontsize=6)
plt.xlabel('Pays')
plt.ylabel('Leviers')
plt.plot([0, 120], [seuil_levier, seuil_levier], color='r')
plt.title("Leviers")
plt.savefig("leviers_reg_3var.jpg")
plt.show()
```

### Projection des résidus studentisés

Le seuil pour les résidus studentisés est une loi de Student à n-p-1 degrés de liberté.

```python
analyse_resultat['rstudent'] = log_reg_multi2.get_influence().resid_studentized_internal
seuil_rstudent = t.ppf(1-alpha/2,n-p-1)
```

```python
plt.figure(figsize=(12,6), dpi=300)
plt.bar(analyse_resultat['country'], analyse_resultat['rstudent'])
plt.xticks(np.arange(0, 12000, step=2), fontsize=5)
plt.xlabel('Pays')
plt.ylabel('Résidus studentisés')
plt.plot([0, 120], [seuil_rstudent, seuil_rstudent], color='r')
plt.plot([0, 120], [-seuil_rstudent, -seuil_rstudent], color='r')
plt.title("Analyse des résidus studentisés")
plt.savefig("rstudent_3var.jpg")
plt.show()
```

### Homoscédasticité

On vérifie que la variance des erreurs de la régression est la même pour chaque observation.
Pour ce faire, on réalise un test de Breusch Pagan, ou on vérifie la valeur de la p-value, au seuil de 5%.

H0 : les variances sont constantes (hypothèse d'homoscédasticité).

```python
_, pval, __, f_pval = sm.stats.diagnostic.het_breuschpagan(log_reg_multi2.resid, col_var2)
print('p value test Breusch Pagan:', pval)
```

La p-value est inférieure à 5%, on rejette donc l'hyptohèse selon laquelle les variances sont constantes.

### Normalité des résidus

On vérifie que les écarts que l’on constate entre les valeurs prédites et les valeurs observées (les résidus donc) sont entièrement imputables à des erreurs de mesure. De ce fait, les résidus doivent posséder les propriétés classiques d’une distribution normale.

```python
shapiro(log_reg_multi2.resid)
```

```python
residuals2 = log_reg_multi2.resid
analyse_resultat['residuals'] = residuals2
analyse_resultat
```

```python
# Représentation graphique de la distribution des résidus

plt.subplots(figsize=(10, 6))
plt.title('Résidus', fontsize=16)
sns.distplot(analyse_resultat['residuals'])

plt.savefig('Résidus')
plt.show()
```

Ici, l'hypothèse de normalité est remise en cause (p-value = 0.003 < 0.05).

Néanmoins, l'observation des résidus, le fait qu'ils ne soient pas très différents d'une distribution symétrique, et le fait que l'échantillon soit de taille suffisante (supérieure à 30) permettent de dire que **les résultats obtenus par le modèle linéaire gaussien ne sont pas absurdes**, même si le résidu n'est pas considéré comme étant gaussien.
